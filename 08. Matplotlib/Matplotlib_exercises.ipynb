### Exercise 1: Read the Total profit of all months and show it using a line plot.
Total profit data provided for each month. The enerated line plot must include the following properties:<br>
- X label name = Month Number
- Y label name = Total profit

The line plot graph should look like this:


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
# Write your code here
url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'

data = pd.read_csv('C:/Users/Varun KR/Downloads/company_sales_data.csv')

data.head()
```

```python
fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(15)

profitList = data['total_profit'].tolist()
monthList  = data['month_number'].tolist()

plt.plot(monthList, profitList, label = 'Monthly Profit')

plt.xlabel('Month number')
plt.ylabel('Profit in Euros')

plt.title('Company profit per month')
plt.show()
```

### Exercise 2: Get the Total profit of all months and show a line plot with the following Style properties:
- Line Style dotted and Line color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read

The line plot graph should look like this:


```python
# Write your code here
fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(15)

profitList = data['total_profit'].tolist()
monthList  = data['month_number'].tolist()

plt.plot(monthList, profitList, label = 'Profit', color='r', marker='o', markerfacecolor='k', linestyle='--', linewidth=3)
      
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.legend(loc='lower right')

plt.title('Company sales data')
plt.show()
```

### Exercise 4: Read the toothpaste sales data of each month and show it using a scatter plot.

The scatter plot should look like this:


```python
# Write your code here

fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(15)

plt.scatter(monthList, ToothPasteSalesData, label = 'Toothpaste Sales data')
plt.xlabel('Month number')
plt.ylabel('Number of units sold')
plt.legend(loc='upper left')
plt.title(' Toothpaste sales data')
plt.grid(True, linewidth= 1, linestyle="--")
plt.show()
```

###Â Exercise 5: Read the face cream and facewash product sales data and show it using the bar chart.

The bar chart should look like this:


```python
# Write your code here

fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(15)

plt.bar([a-0.25 for a in monthList], FaceCremSalesData, width= 0.25, label = 'Face-cream sales data', align='edge')
plt.bar([a+0.25 for a in monthList], FaceWashSalesData, width= -0.25, label = 'Face-wash sales data', align='edge')
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title(' Sales data')


plt.grid(True, linewidth= 1, linestyle="--")
plt.title('Face-wash and Face-cream sales data')
plt.show()
```

### Exercise 8: Calculate the total sale data for last year for each product and show it using a Pie chart.
Note: In the Pie chart you should display the Number of units sold per year for each product in percentage.

The Pie chart should look like this:


```python
# Write your code here

fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(15)

monthList  = data['month_number'].tolist()

labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesData   = [data['facecream'].sum(), data['facewash'].sum(), data['toothpaste'].sum(), data['bathingsoap'].sum(),
                                                                         data['shampoo'].sum(), data['moisturizer'].sum()]
plt.axis("equal")
plt.pie(salesData, labels=labels, autopct='%.2f%%')

plt.legend(loc='lower right')

plt.title('Sales data')
plt.show()

```
