```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
```

1. Plot the absolute value of x and its derivative. Choose an interval to have a nice looking shape.

```python
x = np.linspace(-50,50, 500)
y = abs(x)

plt.plot(x,y)
```

```python
def ab(a):
    return abs(a)

def de(b):
    return derivative(ab, b)


x = np.linspace(-50,50, 500)

plt.plot(x,de(x))
```

2. Plot the function $x\log(x)$ and its derivative.

```python
x = np.linspace(1, 100, 100)
plt.plot(x, x*np.log(x))
```

```python
def l(a):
    return x*np.log(a)

def d(b):
    return derivative(l, b)


x = np.linspace(1, 100, 100)

plt.plot(x,d(x))
```

3. Plot the function $e^{x}\log(x)+5$ and its derivative.

```python
x = np.linspace(1, 100,100)
y = (np.e**x)*np.log(x) + 5

plt.plot(x,y)
```

```python
def l(a):
    return (np.e**x)*np.log(a)

def d(b):
    return derivative(l, b)


x = np.linspace(1, 100, 100)

plt.plot(x,d(x))
```
